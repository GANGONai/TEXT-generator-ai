# 🎶 Song Lyrics AI Studio — Colab launcher

Run the cell below. It will:
1. Clone the project (skipped if already cloned).
2. Install all dependencies.
3. Launch the Gradio web UI with a public `*.gradio.live` URL.

> Tip: switch the runtime to **GPU** (`Runtime → Change runtime type → T4 GPU`) before running for much faster training and generation.

```python
# One-cell launcher for Google Colab.
import os, subprocess, sys, pathlib

REPO = 'https://github.com/GANGONai/TEXT-generator-ai.git'
DIR  = pathlib.Path('/content/TEXT-generator-ai')

if not DIR.exists():
    subprocess.check_call(['git', 'clone', '--depth', '1', REPO, str(DIR)])
else:
    subprocess.check_call(['git', '-C', str(DIR), 'pull', '--ff-only'])

os.chdir(DIR)
subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', '-r', 'requirements.txt'])

os.environ['LYRICS_AI_SHARE'] = '1'
subprocess.check_call([sys.executable, 'app.py', '--share', '--no-browser'])
```
