# 🎶 Song Lyrics AI Studio — Colab launcher

Run the cell below. It will:
1. Clone (or update) the project under `/content/TEXT-generator-ai`.
2. Install all dependencies (~2 minutes on a fresh runtime).
3. Launch the Gradio web UI.

A public `https://*.gradio.live` URL will appear right here in the cell output — open it to use the app.

> Tip: switch the runtime to **GPU** (`Runtime → Change runtime type → T4 GPU`) before running for much faster training and generation.

````python
# One-cell launcher for Google Colab.
# IMPORTANT: we import and call `launch()` directly instead of using
# subprocess so the public *.gradio.live URL prints into this cell.
import os, subprocess, sys, pathlib

REPO = 'https://github.com/GANGONai/TEXT-generator-ai.git'
DIR  = pathlib.Path('/content/TEXT-generator-ai')

if not DIR.exists():
    subprocess.check_call(['git', 'clone', '--depth', '1', REPO, str(DIR)])
else:
    subprocess.check_call(['git', '-C', str(DIR), 'pull', '--ff-only'])

os.chdir(DIR)
if str(DIR) not in sys.path:
    sys.path.insert(0, str(DIR))

# Install deps quietly. `gradio` is the only one not pre-installed in Colab.
subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', '-r', 'requirements.txt'])

# Import and launch — the public *.gradio.live URL will appear below.
from app import launch
launch()  # auto-detects Colab → share=True, inbrowser=False
````
