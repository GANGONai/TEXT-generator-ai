# 🍔 fat_photo — Colab launcher

Это бесплатное приложение добавляет вес человеку на фото. Запусти ячейку ниже — она:
1. Склонирует репозиторий в `/content/TEXT-generator-ai`.
2. Поставит зависимости (`mediapipe`, `diffusers`, `gradio`…), ~2 минуты.
3. Запустит Gradio-приложение.

Публичный URL вида `https://xxxx.gradio.live` появится в выводе ячейки — открой его и пользуйся.

> Перед запуском включи GPU: `Runtime → Change runtime type → T4 GPU`. Без GPU «AI-улучшение» работать будет, но очень медленно (~10 минут на кадр); геометрический warp работает мгновенно и на CPU.

```python
import os, subprocess, sys, pathlib

REPO = 'https://github.com/GANGONai/TEXT-generator-ai.git'
DIR  = pathlib.Path('/content/TEXT-generator-ai')

if not DIR.exists():
    subprocess.check_call(['git', 'clone', '--depth', '1', REPO, str(DIR)])
else:
    subprocess.check_call(['git', '-C', str(DIR), 'pull', '--ff-only'])

os.chdir(DIR / 'fat_photo')
if str(DIR) not in sys.path:
    sys.path.insert(0, str(DIR))

subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', '-r', 'requirements.txt'])

from fat_photo.app import launch
launch()  # auto-share=True on Colab, печатает gradio.live URL ниже
```

## Подсказки

- **Только геометрия (быстро, без GPU)** — оставь чекбокс «AI-улучшение» выключенным. Этого достаточно для большинства фото.
- **AI-улучшение** — поставь галочку. Первый запуск скачает ~2 ГБ весов Stable Diffusion (Realistic Vision 5.1). Дальше — ~30–60 секунд на кадр на T4.
- **Сила деформации** — ползунок 0–100%. Реалистичные значения: 30–60%. Выше 80% — карикатурно.
- **Сохранение фона** — фон автоматически остаётся нетронутым (маска человека от MediaPipe SelfieSegmentation).
- **Лучший результат** — фото в полный рост или поясной портрет одного человека лицом к камере.
